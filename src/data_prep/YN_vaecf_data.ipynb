```python
import os, sys
import pandas as pd
import numpy as np
import random
import pickle
from scipy import sparse
```

```python
new_df = pd.read_csv(os.path.join("..","..","data","TA_User_Reviws_Korea_all_v2_new_df.csv"))
new_df = new_df.drop(['location_name'], axis=1)
new_df.head() 
```

```python
unique_user_ids = new_df.userID.unique()

np.random.seed(1)
np.random.shuffle(unique_user_ids) # to ensure random splitting between train, val and test
unique_location_ids = new_df.locationId.unique()

number_of_users = len(unique_user_ids)
print("total number of users: ", number_of_users)

number_of_locations = len(unique_location_ids)
print("total number of locations: ", number_of_locations)
```

```python
# split users into training, validation and test
val_user_ids = []
test_user_ids = []
train_user_ids = []

for i in range(10000):
    val_user_ids.append(unique_user_ids[i]) # first 10k after shuffling keys

for i in range(10000, 20000):
    test_user_ids.append(unique_user_ids[i]) # next 10k after shuffling keys

for i in range(20000, number_of_users): # all the remaining form training data
    train_user_ids.append(unique_user_ids[i])
```

```python
# creating a location2idId and userId to index dictionary for creating train_data ndarray
location2id = {}
location2id = dict((mid, i) for (i, mid) in enumerate(unique_location_ids))
```

```python
print("creating training data....")

user2id = {}
user2id = dict((uid, i) for (i, uid) in enumerate(train_user_ids))

rows = []
cols = []
for u_id in train_user_ids:
	print("train-",user2id[u_id])
	m_ids = new_df[(new_df.userID == u_id) & (new_df.rating > 3.5)]['locationId'].tolist()
	location_indexes = [location2id[m] for m in m_ids]
	rows.extend([user2id[u_id] for i in range(len(m_ids))])
	cols.extend(location_indexes)
    
    
# creating a sparse matrix with no_of_train_users X movies for training, binarized feedback
# rows and cols should be of same length
train_data = sparse.csr_matrix((np.ones_like(rows),(np.array(rows), np.array(cols))), dtype='float64', shape=(len(train_user_ids), number_of_locations))

# dumping variable to load later for use in VAE code
pickle.dump(train_data, open(os.path.join("..","..","data","train_data.file"), "wb"))
print("number of training users: ", len(train_user_ids))
```

```python
print("creating test data....")
test_user2id = {}
test_user2id = dict((uid, i) for (i, uid) in enumerate(test_user_ids))

test_rows = []
test_cols = []
for u_id in test_user_ids:
	print("test-", test_user2id[u_id])
	m_ids = new_df[(new_df.userID == u_id) & (new_df.rating > 3.5)]['locationId'].tolist()
	location_indexes = [location2id[m] for m in m_ids]
	test_rows.extend([test_user2id[u_id] for i in range(len(m_ids))])
	test_cols.extend(location_indexes)

test_data = sparse.csr_matrix((np.ones_like(test_rows),(np.array(test_rows), np.array(test_cols))), dtype='float64', shape=(len(test_user_ids), number_of_locations))
pickle.dump(test_data, open(os.path.join("..","..","data","test_data.file"), "wb"))
print("number of test users: ", len(test_user_ids))
```

```python
print("creating validation data")
val_user2id = {}
val_user2id = dict((uid, i) for (i, uid) in enumerate(val_user_ids))

val_rows = []
val_cols = []
for u_id in val_user_ids:
	print("val-", val_user2id[u_id])
	m_ids = new_df[(new_df.userID == u_id) & (new_df.rating > 3.5)]['locationId'].tolist()
	location_indexes = [location2id[m] for m in m_ids]
	val_rows.extend([val_user2id[u_id] for i in range(len(m_ids))])
	val_cols.extend(location_indexes)

val_data = sparse.csr_matrix((np.ones_like(val_rows),(np.array(val_rows), np.array(val_cols))), dtype='float64', shape=(len(val_user_ids), number_of_locations))
pickle.dump(val_data, open(os.path.join("..","..","data","val_data.file"), "wb"))
print("number of validation users: ", len(val_user_ids))
```

```python
all_user_ids = []

for i in range(0, number_of_users): # all the remaining 
    all_user_ids.append(unique_user_ids[i])
```

```python
print("creating all data")
all_user2id = {}
all_user2id = dict((uid, i) for (i, uid) in enumerate(all_user_ids))

all_rows = []
all_cols = []
for u_id in all_user_ids:
	print("all-", all_user2id[u_id])
	m_ids = new_df[(new_df.userID == u_id) & (new_df.rating > 3.5)]['locationId'].tolist()
	location_indexes = [location2id[m] for m in m_ids]
	all_rows.extend([all_user2id[u_id] for i in range(len(m_ids))])
	all_cols.extend(location_indexes)

all_data = sparse.csr_matrix((np.ones_like(all_rows),(np.array(all_rows), np.array(all_cols))), dtype='float64', shape=(len(all_user_ids), number_of_locations))
pickle.dump(all_data, open(os.path.join("..","..","data","all_data.file"), "wb"))
print("number of allidation users: ", len(all_user_ids))
```
